**Sign Language Gesture Images Dataset**

**Problem Statement:**

Sign language is a vital communication method for individuals with hearing and speech impairments, but communication with non-sign language users remains a significant challenge. The lack of readily available interpreters often creates barriers in education, healthcare, workplaces, and public services. This project aims to develop a Sign Language Gesture Recognition System using Computer Vision and Machine Learning to accurately classify hand gestures from images, enabling faster, more accessible, and inclusive communication.

**Objective**


The objective of this project is to develop a Sign Language Gesture Recognition System using Computer Vision and Machine Learning. The system is designed to recognize and classify hand gestures from images accurately. It also aims to evaluate multiple machine learning algorithms and select the best-performing model for reliable gesture recognition, improving communication accessibility for individuals with hearing and speech impairments.


**Classification Dataset**: Sign Language Gesture Images Dataset

**Source:Kaggle**

**Target Classes:**

The target classes are the categories that the machine learning model is trained to predict. In this dataset, the target classes consist of 37 gesture categories, including the English alphabets (A to Z), numeric digits (0 to 9), and the '-' (hyphen) gesture. Each image is assigned the label corresponding to the folder in which it is stored.

**Input Images:**

The input consists of preprocessed sign language gesture images, where each image contains a hand performing a specific sign. These images are provided to the machine learning model, which extracts hand features using Computer Vision techniques and predicts the corresponding gesture class.

**Business Objective**

The objective of this project is to develop a reliable and efficient Sign Language Gesture Recognition System that automatically identifies hand gestures from images using Computer Vision and Machine Learning. The system aims to improve communication accessibility for individuals with hearing and speech impairments by providing accurate gesture recognition. It also seeks to support real-world applications in education, healthcare, customer service, and public services by enabling faster and more effective interaction between sign language users and non-sign language users.

```python
#Normalize All Images
import os
import cv2
import numpy as np

images = []
labels = []

for label in sorted(os.listdir(dataset_path)):
    class_path = os.path.join(dataset_path, label)

    if os.path.isdir(class_path):
        for file in os.listdir(class_path):

            img_path = os.path.join(class_path, file)

            image = cv2.imread(img_path)

            if image is not None:

                # Normalize pixel values (0–255 → 0–1)
                image = image.astype(np.float32) / 255.0

                images.append(image)
                labels.append(label)

print("Images Normalized Successfully!")
print("Total Images:", len(images))
```

# **Step 4:** Feature Extraction


Classical Machine Learning

Extract image features using

HOG (Histogram of Oriented Gradients)


```python
from skimage.feature import hog
import numpy as np
import os
import cv2
```

```python
from skimage.feature import hog
import os
import cv2
import numpy as np

X = []
y = []

IMG_SIZE = (50, 50)

for label in sorted(os.listdir(dataset_path)):
    class_path = os.path.join(dataset_path, label)

    if os.path.isdir(class_path):
        for file in os.listdir(class_path):

            img_path = os.path.join(class_path, file)

            image = cv2.imread(img_path)

            if image is not None:

                # Resize
                image = cv2.resize(image, IMG_SIZE)

                # Convert to grayscale
                gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

                # Normalize
                gray = gray.astype(np.float32) / 255.0

                # Extract HOG features
                features = hog(
                    gray,
                    orientations=9,
                    pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2),
                    block_norm='L2-Hys'
                )

                X.append(features)
                y.append(label)

# Convert to NumPy arrays
X = np.array(X)
y = np.array(y)

print("Feature Matrix Shape:", X.shape)
print("Labels Shape:", y.shape)
```

visualize HOG

```python
import os
import cv2
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import exposure

# Read one sample image
sample_class = sorted(os.listdir(dataset_path))[0]
sample_image = os.listdir(os.path.join(dataset_path, sample_class))[0]

img_path = os.path.join(dataset_path, sample_class, sample_image)

image = cv2.imread(img_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Extract HOG features and HOG image
features, hog_image = hog(
    gray,
    orientations=9,
    pixels_per_cell=(8, 8),
    cells_per_block=(2, 2),
    block_norm='L2-Hys',
    visualize=True
)

# Improve HOG image contrast for display
hog_image = exposure.rescale_intensity(hog_image, in_range='image')

# Display
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(hog_image, cmap='gray')
plt.title("HOG Feature Visualization")
plt.axis("off")

plt.show()
```

# **Step 7:** Model Building


```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score
```

Logistic Regression

```python
lr = LogisticRegression(random_state=42, max_iter=1000)

lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)

lr_accuracy = accuracy_score(y_test, y_pred_lr)

print("Logistic Regression Accuracy:", round(lr_accuracy,4))
```

Decision Tree

# **Step 10:** Final Conclusion


Project Workflow

The project began by importing the required Python libraries and loading the image dataset from Google Drive. Dataset exploration (EDA) was performed to understand the number of images, class labels, image resolution, image format, class distribution, image dimensions, color channels, and overall image quality. Several visualizations, including class count plots, sample images, and image resolution distributions, were created to gain insights into the dataset.

Since the dataset was already preprocessed, only minimal preprocessing was required. All images were verified to have the same dimensions (50 × 50 pixels), pixel values were normalized to the range 0 to 1, and the images were converted from RGB to grayscale to make them suitable for Histogram of Oriented Gradients (HOG) feature extraction. The dataset was also checked to ensure there were no corrupted images.

Feature extraction was performed using Histogram of Oriented Gradients (HOG), which captures the shape and edge information of hand gestures. Each image was converted into a numerical feature vector containing 900 HOG features, which served as the input for the machine learning models.

The extracted features (X) and corresponding class labels (y) were separated and the dataset was divided into 80% training data and 20% testing data using stratified train-test splitting with random_state=42. Stratification ensured that all 37 gesture classes maintained similar distributions in both the training and testing datasets.

Several baseline machine learning algorithms were trained, including Logistic Regression, Decision Tree, Random Forest, K-Nearest Neighbors (KNN), and Gaussian Naive Bayes. Their performances were compared using standard classification metrics.

After comparing the baseline models, Random Forest was selected as the best-performing algorithm. Hyperparameter tuning was performed using GridSearchCV to optimize important parameters such as the number of trees, maximum tree depth, minimum samples required for splitting, and minimum samples required at leaf nodes. The tuned Random Forest model achieved improved classification performance.

The final model was evaluated using Accuracy, Precision, Recall, F1-Score, Classification Report, and Confusion Matrix. Since the project involves 37 gesture classes, weighted averages were used for Precision, Recall, and F1-Score to provide an overall evaluation across all classes. ROC-AUC was not included because it is primarily intended for binary classification problems.

Finally, the trained Random Forest model was prepared for deployment by saving it using Joblib/Pickle. The saved model can be integrated with MediaPipe Hands and OpenCV to perform real-time sign language gesture recognition using a webcam. The system detects hand movements, processes the captured gesture, extracts HOG features, and predicts the corresponding gesture class, displaying the recognized symbol on the live video feed.

By combining Computer Vision, HOG feature extraction, and a tuned Random Forest classifier, the system provides reliable gesture recognition and forms a strong foundation for real-time sign language applications. The solution demonstrates the practical use of machine learning in improving accessibility, promoting inclusive communication, and supporting assistive technologies for individuals with hearing and speech impairments.
